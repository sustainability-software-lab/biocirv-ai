# BioCirv AI-Driven Data Analysis

This notebook demonstrates how to use the `ca_biositing.ai_exploration` package to perform natural language analysis across multiple materialized views in the CA Biositing project. 

It leverages **PandasAI 3.0+** and the **CBORG API** (Gemini models) at LBL.

### Step 1: Setup and Environment

We start by initializing the sandbox. The `init_sandbox` function handles LLM configuration and database connectivity.

**Note:** Ensure you have selected the **Python (BioCirv AI)** kernel.

```python
from ca_biositing.ai_exploration.sandbox_setup import init_sandbox, get_agent

llm, engine = init_sandbox(model_name="gemini-2.0-flash")
```

### Step 2: Initialize AI Agent

We can initialize the agent by specifying views or let it automatically discover all available views in the `ca_biositing` and `analytics` schemas.

```python
# Automatic view discovery
agent = get_agent(llm, engine)

print("Agent initialized with dynamic schema discovery and custom SQL skill.")
```

### Step 3: Natural Language Queries

Now we can query the data using plain English. 

```python
agent.chat("Which are the top 5 resources by volume in California according to the analysis data?")
```

### Step 4: Visualizations

You can also ask the AI to generate interactive charts.

```python
agent.chat("Show me an interactive plotly bar chart of total biomass volume by resource type")
```
